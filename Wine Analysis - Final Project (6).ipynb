# Phase 4: Predictive Modeling

## Identifying High-Quality Wines

The exploratory analysis identified several chemical characteristics that appear to be associated with wine quality. While these findings provide valuable insight, **they do not indicate how effectively wine quality can be predicted using physicochemical measurements alone**.

To further evaluate these relationships, a predictive modeling approach was implemented. The dataset was separated into red and white wines to determine whether the **factors associated with high-quality wines differ between wine types**.

The objective of this section is to assess the ability of chemical characteristics to predict whether a wine will be classified as high quality and identify the variables that contribute most strongly to those predictions.


## Model Development

A **logistic regression model** was selected as the initial classification approach due to its **interpretability and effectiveness for binary classification problems**.

The modeling process included:

1. Separating predictor variables from the target variable.
2. Splitting the dataset into training and testing subsets.
3. Standardizing numerical features to ensure consistent scaling.
4. Training a logistic regression classifier.
5. Evaluating model performance using accuracy, precision, recall, and confusion matrices.

This process was performed independently for red and white wines to determine whether predictive performance differed between wine types.


```python
region["high_quality"] = np.where(region["quality"] >= 7, 1, 0)

region[["quality", "high_quality"]].head()
```

```python
region["high_quality"].value_counts(normalize=True)
```

```python
red_region = region[region["wine_type"] == "R"].copy()
white_region = region[region["wine_type"] == "W"].copy()

print(red_region.shape)
print(white_region.shape)
```

## Model Performance: Red Wine Classification

The logistic regression model demonstrated a strong ability to distinguish between high-quality and non-high-quality red wines using physicochemical characteristics alone.

The **confusion matrix and classification metrics** indicate that the **model successfully identified a substantial proportion of wines in each category while maintaining overall predictive consistency**. 

This suggests that **measurable chemical properties contain meaningful information** related to **perceived wine quality**.

Although the model is not perfect, the results demonstrate that **wine quality is not entirely subjective** and can be **partially explained through objective laboratory measurements**. 

The performance achieved by the model supports the conclusion that several of the variables identified during exploratory analysis possess meaningful predictive value.

Future modeling approaches such as Random Forest, Gradient Boosting, or XGBoost may improve predictive performance by capturing more complex relationships among variables.


```python
drop_cols = ["quality", "high_quality", "wine_type", "wine_quality"]

X_red = red_region.drop(drop_cols, axis=1)
y_red = red_region["high_quality"]

X_white = white_region.drop(drop_cols, axis=1)
y_white = white_region["high_quality"]
```

```python
X_red_train, X_red_test, y_red_train, y_red_test = train_test_split(
    X_red,
    y_red,
    test_size=0.25,
    random_state=42,
    stratify=y_red
)

scaler_red = StandardScaler()

X_red_train_scaled = scaler_red.fit_transform(X_red_train)
X_red_test_scaled = scaler_red.transform(X_red_test)

red_model = LogisticRegression(max_iter=5000)
red_model.fit(X_red_train_scaled, y_red_train)

red_pred = red_model.predict(X_red_test_scaled)

print(confusion_matrix(y_red_test, red_pred))
print(classification_report(y_red_test, red_pred))
print("Red Wine Accuracy:", accuracy_score(y_red_test, red_pred))
```

## White Wine Predictive Model

To determine whether predictive relationships differ across wine types, the same modeling process was applied to the white wine dataset. Comparing model performance between red and white wines provides additional insight into the consistency of quality drivers across wine categories.

```python
X_white_train, X_white_test, y_white_train, y_white_test = train_test_split(
    X_white,
    y_white,
    test_size=0.25,
    random_state=42,
    stratify=y_white
)

scaler_white = StandardScaler()

X_white_train_scaled = scaler_white.fit_transform(X_white_train)
X_white_test_scaled = scaler_white.transform(X_white_test)

white_model = LogisticRegression(max_iter=5000)
white_model.fit(X_white_train_scaled, y_white_train)

white_pred = white_model.predict(X_white_test_scaled)

print(confusion_matrix(y_white_test, white_pred))
print(classification_report(y_white_test, white_pred))
print("White Wine Accuracy:", accuracy_score(y_white_test, white_pred))
```

## Feature Importance Analysis

While overall model accuracy indicates how well the logistic regression models classify high-quality wines, it does not explain which characteristics are driving those predictions.

To better understand the factors associated with high-quality wines, the **coefficients from the red and white wine logistic regression models were compared**. 

**Positive coefficients** indicate variables that increase the likelihood of a wine being classified as **high quality**, while **negative coefficients** indicate variables associated with **lower-quality classifications**.

Comparing coefficient magnitudes also provides insight into which physicochemical characteristics have the strongest influence on model predictions for each wine type.


```python
red_coefficients = pd.DataFrame({
    "Feature": X_red.columns,
    "Red_Coefficient": red_model.coef_[0]
}).sort_values(by="Red_Coefficient", ascending=False)

white_coefficients = pd.DataFrame({
    "Feature": X_white.columns,
    "White_Coefficient": white_model.coef_[0]
}).sort_values(by="White_Coefficient", ascending=False)

comparison = pd.merge(
    red_coefficients,
    white_coefficients,
    on="Feature",
    how="outer"
)

comparison["Red_Abs_Importance"] = comparison["Red_Coefficient"].abs()
comparison["White_Abs_Importance"] = comparison["White_Coefficient"].abs()

comparison
```

For **white wines**, **residual sugar** demonstrated one of the **strongest positive relationships** with high quality, followed by pH and fixed acidity. **Density emerged as the strongest negative predictor**, suggesting that lower-density white wines were more likely to be classified as high quality. Volatile acidity also maintained a negative relationship with quality, consistent with the findings observed for red wines.

Several variables exhibited similar directional effects across both wine types. **Alcohol content was positively associated with quality in both models**, while **volatile acidity and chlorides were negatively associated with quality**. These consistent patterns suggest that certain physicochemical characteristics may serve as reliable indicators of wine quality regardless of wine type.

Overall, the results indicate that while red and white wines share some common quality drivers, **important differences exist in the relative influence of specific chemical characteristics**. This supports the decision to model red and white wines separately and highlights the complexity of factors that contribute to perceived wine quality.


### Key Takeaways

Unlike logistic regression, Random Forest can capture nonlinear interactions between variables and does not require assumptions regarding the form of those relationships.


```python
# Random Forest Validation Model

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix
)

# Recreate predictors and target variables if needed
drop_cols = ["quality", "high_quality", "wine_type", "wine_quality"]

X_red = red_region.drop(drop_cols, axis=1, errors="ignore")
y_red = red_region["high_quality"]

X_white = white_region.drop(drop_cols, axis=1, errors="ignore")
y_white = white_region["high_quality"]


def run_random_forest_validation(X, y, wine_label):
    """
    Trains and validates a Random Forest model for high-quality wine classification.
    """

    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size=0.25,
        random_state=42,
        stratify=y
    )

    rf_model = RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        class_weight="balanced",
        max_depth=None,
        min_samples_split=2,
        min_samples_leaf=1
    )

    rf_model.fit(X_train, y_train)

    y_pred = rf_model.predict(X_test)
    y_prob = rf_model.predict_proba(X_test)[:, 1]

    print(f"\n{wine_label} Wine Random Forest Validation Results")
    print("-" * 55)
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    print("Accuracy:", round(accuracy_score(y_test, y_pred), 4))
    print("Precision:", round(precision_score(y_test, y_pred), 4))
    print("Recall:", round(recall_score(y_test, y_pred), 4))
    print("F1 Score:", round(f1_score(y_test, y_pred), 4))
    print("ROC-AUC:", round(roc_auc_score(y_test, y_prob), 4))
    
     # 5-fold cross-validation
    scoring = {
        "accuracy": "accuracy",
        "precision": "precision",
        "recall": "recall",
        "f1": "f1",
        "roc_auc": "roc_auc"
    }

    cv_results = cross_validate(
        rf_model,
        X,
        y,
        cv=5,
        scoring=scoring
    )

    cv_summary = pd.DataFrame({
        "Metric": ["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC"],
        "Cross Validation Mean": [
            cv_results["test_accuracy"].mean(),
            cv_results["test_precision"].mean(),
            cv_results["test_recall"].mean(),
            cv_results["test_f1"].mean(),
            cv_results["test_roc_auc"].mean()
        ],
        "Cross Validation Std": [
            cv_results["test_accuracy"].std(),
            cv_results["test_precision"].std(),
            cv_results["test_recall"].std(),
            cv_results["test_f1"].std(),
            cv_results["test_roc_auc"].std()
        ]
    })

    print("\n5-Fold Cross-Validation Summary:")
    display(cv_summary)

    feature_importance = pd.DataFrame({
        "Feature": X.columns,
        "Importance": rf_model.feature_importances_
    }).sort_values(by="Importance", ascending=False)

    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=feature_importance,
        x="Importance",
        y="Feature",
        palette="viridis"
    )
    plt.title(f"{wine_label} Wine Random Forest Feature Importance")
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

    return rf_model, cv_summary, feature_importance


red_rf_model, red_rf_cv, red_rf_importance = run_random_forest_validation(
    X_red,
    y_red,
    "Red"
)

white_rf_model, white_rf_cv, white_rf_importance = run_random_forest_validation(
    X_white,
    y_white,
    "White"
)

```

### Interpretation
Random Forest was used as an independent validation technique to evaluate whether nonlinear relationships would identify different quality drivers. 

The model was **evaluated using a train-test split and 5-fold cross-validation**. Accuracy, precision, recall, F1 score, and ROC-AUC were reviewed to assess performance. Feature importance values were also examined to identify which chemical characteristics contributed most strongly to high-quality wine predictions.

Several variables including alcohol content, volatile acidity, density, sulphates, and residual sugar—continued to rank among the most influential predictors of wine quality.

## Preparing the Regional Analysis Dataset

Unlike the UCI wine-quality dataset used in the physicochemical analysis, the Wine Enthusiast review dataset was not structured specifically for regional analysis. 

Several preprocessing steps were required to **identify Vinho Verde wines, standardize geographic information, and prepare the data** for subregional comparison.

### Regional Dataset Preparation

The following transformations were performed:

* Filtered the review dataset to wines originating from Portugal.
* Identified Vinho Verde wines using region names and subregional references contained within review text and metadata fields.
* Standardized review scores using a common regional scoring field.
* Created a high-quality review indicator for wines receiving review scores of 90 points or higher.
* Classified wines into recognized Vinho Verde subregions using text-based matching rules.
* Inferred wine type (red or white) when a direct classification was not available in the review dataset.

These preparation steps created a regional analysis dataset capable of comparing review performance across Vinho Verde subregions and wine styles.


```python
# Load external wine-review data for regional analysis
wine_reviews = pd.read_csv("winemag-data-130k-v2.csv")

required_cols = ["country", "points"]
missing_cols = [col for col in required_cols if col not in wine_reviews.columns]

if missing_cols:
    raise ValueError("Missing required columns: " + str(missing_cols))

portugal_reviews = wine_reviews[wine_reviews["country"] == "Portugal"].copy()

review_text_cols = [
    col for col in ["province", "region_1", "region_2", "title", "description", "variety", "winery"]
    if col in portugal_reviews.columns
]

import unicodedata

def normalize_text(value):
    value = str(value)
    value = unicodedata.normalize("NFKD", value)
    value = "".join(char for char in value if not unicodedata.combining(char))
    return value.lower()

vinho_verde_terms = [
    "vinho verde", "moncao", "melgaco", "lima", "amarante",
    "baiao", "basto", "cavado", "ave", "paiva", "sousa"
]

review_search_text = portugal_reviews[review_text_cols].fillna("").agg(" ".join, axis=1)
review_search_text = review_search_text.apply(normalize_text)

vinho_verde_reviews = portugal_reviews[
    review_search_text.apply(lambda text: any(term in text for term in vinho_verde_terms))
].copy()

print("Portugal reviews:", portugal_reviews.shape)
print("Vinho Verde reviews:", vinho_verde_reviews.shape)
vinho_verde_reviews.head()
```

### Dataset Filtering Results

The Wine Enthusiast review dataset was first filtered to wines originating from Portugal. A **text-based matching process** was then used to identify reviews associated with Vinho Verde and its recognized subregions.

This process reduced the dataset from 5,691 Portuguese wine reviews to 1,039 reviews specifically associated with Vinho Verde. The resulting dataset serves as the foundation for the regional analysis that follows.


## Comparing Review Performance Across Vinho Verde Subregions

With the regional dataset prepared, the next step is to examine how wine quality varies across the Vinho Verde region.

Review scores were aggregated at the subregional level to compare:

* Number of reviewed wines
* Average review score
* Median review score
* Highest recorded review score

This analysis provides an initial view of whether certain subregions consistently produce higher-rated wines than others.


```python
# Standardize regional-review fields
subregion_data = vinho_verde_reviews.copy()
subregion_data = subregion_data.rename(columns={"points": "regional_score"})

# Use review points as the high-quality definition for the regional dataset.
# This is separate from the UCI 7+ quality scale.
subregion_data["high_quality_review"] = np.where(subregion_data["regional_score"] >= 90, 1, 0)

subregion_data[["regional_score", "high_quality_review"]].head()
```

```python
# Identify Vinho Verde subregion from available text fields
import unicodedata

def normalize_text(value):
    value = str(value)
    value = unicodedata.normalize("NFKD", value)
    value = "".join(char for char in value if not unicodedata.combining(char))
    return value.lower()

subregion_aliases = {
    "Moncao e Melgaco": ["moncao e melgaco", "moncao", "melgaco"],
    "Lima": ["lima"],
    "Cavado": ["cavado"],
    "Ave": ["ave"],
    "Basto": ["basto"],
    "Amarante": ["amarante"],
    "Baiao": ["baiao"],
    "Paiva": ["paiva"],
    "Sousa": ["sousa"]
}

def identify_subregion(row):
    text = " ".join([
        str(row.get(col, ""))
        for col in ["province", "region_1", "region_2", "title", "description", "variety", "winery"]
    ])
    text = normalize_text(text)

    for subregion, aliases in subregion_aliases.items():
        if any(alias in text for alias in aliases):
            return subregion

    return "Unspecified Vinho Verde"

subregion_data["subregion"] = subregion_data.apply(identify_subregion, axis=1)
subregion_data["subregion"].value_counts()
```

```python
# Infer wine type when the review dataset does not provide a direct red/white field
white_terms = [
    "white", "branco", "alvarinho", "loureiro", "trajadura",
    "arinto", "avesso", "azal"
]

red_terms = [
    "red", "tinto", "vinhao", "espadeiro", "borracal"
]

def infer_wine_type(row):
    text = " ".join([
        str(row.get(col, ""))
        for col in ["title", "description", "variety", "designation"]
    ])
    text = normalize_text(text)

    if any(term in text for term in red_terms):
        return "R"
    if any(term in text for term in white_terms):
        return "W"
    return "Unspecified"

if "wine_type" not in subregion_data.columns:
    subregion_data["wine_type"] = subregion_data.apply(infer_wine_type, axis=1)

subregion_data["wine_type"].value_counts()
```

```python
# Subregion-level quality summary
subregion_quality = subregion_data.groupby("subregion")["regional_score"].agg(
    count="count",
    avg_score="mean",
    median_score="median",
    max_score="max"
).sort_values(by="avg_score", ascending=False)

subregion_quality
```

### Interpretation

Several differences emerge when comparing review performance across Vinho Verde subregions.

**Sousa** achieved the highest average review score, **exceeding 90 points on average** across reviewed wines. Monção e Melgaço and Basto also performed strongly, with average scores above 88 points. Larger subregions such as Ave and Lima demonstrated slightly lower average scores but contributed substantially more reviewed wines.

Review **volume should be considered when interpreting these results**. Some subregions contain relatively small sample sizes, meaning average scores may be more sensitive to individual reviews. Nevertheless, the **findings suggest that regional variation** exists within Vinho Verde and that certain subregions may produce wines that consistently receive stronger critical evaluations.

These results provide an **initial indication that geography may influence wine quality** outcomes, **complementing the physicochemical findings**.

## Integrating Quality, Geography, and Wine Type

The previous visualizations established that wine quality varies across Vinho Verde subregions and that performance patterns differ between red and white wines. While static charts provide useful comparisons, interactive mapping allows these regional relationships to be explored in greater detail.

The interactive map combines multiple dimensions of the analysis into a single visualization:

- Geographic location of each subregion
- Review volume
- Average review score
- Median and maximum review scores
- Red wine performance metrics
- White wine performance metrics
- Proportion of wines receiving scores of 90 points or higher

Marker size represents review volume, while marker color reflects average review score. Selecting a marker reveals detailed performance information for each subregion.

This visualization serves as the final integration point between the physicochemical quality analysis performed in Phase 1 and the regional review analysis conducted in Phase 2.


## How to Read the Map

This interactive map visualizes regional review performance across the Vinho Verde region.

* Circle color represents average review score.
* Circle size represents the number of reviewed wines.
* Popup windows provide additional review statistics, including red and white wine performance where available.

The color scale ranges from lower average review scores to higher average review scores, while larger circles indicate greater review volume. Combining these measures helps distinguish meaningful regional performance patterns from results based on only a small number of reviews.


```python
import folium
import branca.colormap as cm

def format_score(value):
    if pd.notnull(value):
        return f"{value:.2f}"
    return "N/A"

def format_rate(value):
    if pd.notnull(value):
        return f"{value:.2%}"
    return "N/A"

import pandas as pd
import numpy as np
import folium
import branca.colormap as cm

# Color scale based on average review score
min_score = map_data["avg_score"].min()
max_score = map_data["avg_score"].max()

color_scale = cm.LinearColormap(
    colors=["#d73027", "#fee08b", "#1a9850"],
    vmin=min_score,
    vmax=max_score,
    caption="Average Review Score"
)

vinho_map = folium.Map(
    location=[41.45, -8.25],
    zoom_start=9,
    tiles="OpenStreetMap"
)

for _, row in map_data.iterrows():
    red_avg = row.get("avg_score_R", np.nan)
    white_avg = row.get("avg_score_W", np.nan)
    red_hq = row.get("high_quality_rate_R", np.nan)
    white_hq = row.get("high_quality_rate_W", np.nan)

    marker_color = color_scale(row["avg_score"])

    popup_text = f"""
    <b>{row['subregion']}</b><br>
    Review count: {row['count']:.0f}<br>
    Avg review score: {row['avg_score']:.2f}<br>
    Median review score: {row['median_score']:.2f}<br>
    Max review score: {row['max_score']:.0f}<br><br>
    Red avg score: {format_score(red_avg)}<br>
    White avg score: {format_score(white_avg)}<br>
    Red 90+ rate: {format_rate(red_hq)}<br>
    White 90+ rate: {format_rate(white_hq)}
    """

    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=max(6, min(24, row["count"] * 2)),
        popup=folium.Popup(popup_text, max_width=350),
        tooltip=f"{row['subregion']}: {row['avg_score']:.2f}",
        color="black",
        weight=1,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.75
    ).add_to(vinho_map)

color_scale.add_to(vinho_map)

vinho_map
```
